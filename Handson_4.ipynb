## 6.1 Prime Number Checker
Create a program that checks whether a number is a prime number and displays its factors if it is not a prime number.

### Console:
```powershell

It has 4 factors: 1 2 3 6

Try again? (y/n): y

Please enter an integer between 1 and 5000: 200
200 is NOT a prime number.
It has 12 factors: 1 2 4 5 8 10 20 25 40 50 100 200

Try again? (y/n): n

Bye!
```

### Specifications:
- A prime number is divisible by two factors (1 and itself). For example, 7 is a prime number because it is only divisible by 1 and 7.
- If the user enters an integer that's not between 1 and 5000, the program should display an error message.
- If the number is a prime number, the program should display a message.
If the number is not a prime number, the program should display a message. Then, it should display the number of factors for the number and a list of those factors.
- Store the factors for each number in a list.
- Use functions to organize the code for this program.


````python
def is_prime_num(number):
    if number <= 1:
        print(str(number), " is not a prime number.")
    else:
        for i in range(2, int(number/2)+1):
            if number % i == 0:
                print(str(number), " is not a prime number.")
                factors(number)
                break
        else:
            print(str(number), " is a prime number.")

def factors(number):
    count = []
    if number % 1 == 0 and number % number == 0:
        for i in range(1, number+1):
            if number % i == 0:
                count.append(i)

        if len(count) > 0:
            print("It has " + str(len(count)) + " factors:", end=" ")
            for i in range(len(count) - 1):
                print(count[i], end=", ")
            print(count[-1])
        else:
            print("It has 0 factors.")

def user_input():
    number = int(input("Please enter an integer between 1 and 5000: "))
    return number

def main():
    print("Prime Number Checker")
    print()

    while True:
        number = user_input()

        if 1 <= number and number <= 5000:
            is_prime_num(number)
        else:
            print("Error! Please enter a number between 1 and 5000. ")

        print()
        choice = input("Try again? (y/n): ")
        print()
        if choice.lower() != 'y':
            print()
            print("Bye!")
            print()
            break

if __name__ == "__main__":
    main()
````

## 6.2 - Contact Manager
Create a program that a user can use to manage the primary email address and phone number for a contact.

### Console
```powershell

COMMAND MENU
list - Display all contacts
view - View a contact
add - Add a contact
del - Delete a contact

Command: list

2. Eric Idle

Command: view

Name: Eric Idle

Command: add

Mike Murach was added.

Command: list
1. Eric Idle
2. Mike Murach

```

### Specifications:
- Use a list of lists to store the data for the contacts. Provide starting data for two or more contacts.
- For the `view` and `del` commands, display an error message if the user enters and invalid contact number.
- When you exit the program, all changes that you made to the contact list are lost.




````python
def Contact_list(contacts):
    
    i = 1
    for row in contacts:
        print(str(i)+ "."+ str(row[0]))
        i += 1
    
def view_contact(contacts):
    
    number = int(input("Number: "))
    
    if number < 1 or number > len(contacts):
        print("Error! Invalid contact number.")
        
    else:
        contact_list = contacts[number-1]
        print("Name:", contact_list[0])
        print("Email:", contact_list[1])
        print("Phone:", contact_list[2])
    
def add_contact(contacts):
    
    Name = input("Enter Name:")
    Email = input("Enter E-mail: ")
    Phone = input("Enter Phone: ")
    
    contact_list = [Name, Email, Phone]
    contacts.append(contact_list)
    print(contact_list[0] + " was added.")
    
def del_contact(contacts):
    
    
    number = int(input("Enter number: "))
    if number<1 or number>len(contacts):
        print("Error! input valid integer.")
     
    else:
        deleted_contact = contacts.pop(number - 1)
        print(deleted_contact[0] + " is deleted.")
    
    
def display_menu():
    print("Contact Manager")
    print()
    print("COMMAND MENU")
    print("list - Display all contacts")
    print("view - View a contact")
    print("add - Add a contact")
    print("del - Delete a contact")
    print("exit - Exit program")
    print()
    
def main():
    
    contacts = [["Guido van Rossum"],
                ["Eric Idle", "eric@ericidle.com", "+44 20 7946 0958"]]
    
    display_menu()
    
    while True:
        command = input("Command: ")
        
        if command == "list":
            Contact_list(contacts)
            print()
        elif command == "view":
            view_contact(contacts)
            print()
        elif command == "add":
            add_contact(contacts)
            print()
        elif command == "del":
            del_contact(contacts)
            print()
        elif command == "exit":
            break
        elif command.isdigit():  # Check if input is a number
            view_contact(contacts)
            
        else:
            print("Not a valid command. Please try again.")
            
    print("Bye!")
    
if __name__ == "__main__":
    main()
````
