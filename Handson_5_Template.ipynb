## 7.2 Wizard Inventory
Create a program that keeps track of the items that a wizard can carry.

### Console:
```powershell

```

### Specifications:
- Use the attached text file named `wizard_all_items.txt` that contains a list of all the items that a wizard can carry. (in the corresponding directory)
- When the user selects the walk command, the program should read the items from the file, randomly pick one, and give the user the option to grab it.
- Your program should create another file that stores the items that the wizard is carrying (`wizard_inventory.txt`). Make sure to update this file every time the user grabs or drops an item.
- The wizard can only carry **four** items at a time.
- For the drop command, display an error message if the user enters an invalid number for the item.



````python
import random

def read_items(filename): 
    
    with open(filename, "r") as file:
        items = file.readlines()
    return items
        
    
def show_inventory(inventory):
    
    index = 1
    for item in inventory:
            print(str(index) + ". " + item.replace('\n', ''))
            index += 1

def update_inventory(filename, inventory): 
    
    with open(filename, "w")as file:
        for item in inventory:
            file.write(item + "\n")
    

def pick_up_items(inventory,items_file,max_capacity):
    
    items = read_items(items_file)

    while True:
        random_item = random.choice(items).strip()
        if random_item not in inventory:
            break

    print(f"While walking down a path, you see {random_item}.")
    user_input = input("Do you want to grab it? (y/n): ")

    if user_input.lower() == "y":
        if len(inventory) >= max_capacity:
            print("You can't carry any more items. Drop something first.")
            return

        inventory.append(random_item)
        print(f"You picked up {random_item}")
    
    
def main():
    max_capacity = 4
    inventory = []
    
    with open("wizard_inventory.txt", "r") as file:
        inventory = file.readlines()
    
    print("The Wizard Inventory Program")
    print()
    print("COMMAND MENU")
    print("walk - Walk down the path ")
    print("show - Show all items")
    print("drop - Drop an item")
    print("exit - Exit program")
    print()
    
    while True:
        choice = input("Command: ")
       
        if choice.lower() == "walk":
            
            picked_item = pick_up_items(inventory,"wizard_all_items.txt",max_capacity)
            if picked_item:
                
                if len(inventory) <= max_capacity:
                    user_input = input("Do you want to grab it? (y/n): ")
                    if user_input.lower() == "y":
                        print("You picked up ", picked_item)
                    else:
                        continue
                    
                else:
                    continue
            
          
            print()
        elif choice.lower() == "show":
            show_inventory(inventory)
            print()
        elif choice.lower() == "drop":
            show_inventory(inventory)
            item_number = input("Drop number: ")
            if item_number.isdigit() and 1 <= int(item_number) <= len(inventory):
                dropped_item = inventory.pop(int(item_number) - 1)
                print("You dropped", dropped_item.replace('\n', ''))
                update_inventory("wizard_inventory.txt", inventory)
            else:
                print("Invalid item number.")
            
            print()
                
        elif choice.lower() == "exit":
            update_inventory("wizard_inventory.txt", inventory)
            print("Bye!")
            break
            
        else:
            print("Invalid command. Please try again.")
                
if __name__=="__main__":
    main()

            
    
    
````
